## Importing Libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(18, 8))
sns.countplot(data = df, x = "category")
plt.xlabel("Product Category")
plt.ylabel("Count of Products")
plt.xticks(rotation = 45)
plt.title("Product Categories missing ratings")
```

```python
sns.catplot(data = df.sort_values(by = "rating", ascending=False), x = "category", y = "rating", kind = "boxen", height = 7, aspect = 3)
plt.xlabel("Product Category")
plt.ylabel("rating of Products")
plt.xticks(rotation = 45)
plt.title("Product Categories and their rating Distribution")
```

```python
sns.histplot(data = df, y = "sale_price", x = "rating", hue = "category", bins=5, alpha = 0.7)
```

```python
sns.catplot(data = df.sort_values(by = "sale_price", ascending=False), x = "category", y = "sale_price", kind = "boxen", height=7, aspect=3)
plt.xlabel("Product Category")
plt.ylabel("Price of Products")
plt.xticks(rotation = 45)
plt.title("Product Categories and their Sale Price Distribution")
```

```python
df.sort_values("sale_price", ascending= False)
```

```python
plt.figure(figsize = (8, 6))
plt.rcParams["figure.figsize"] = (18,5.5)
df.groupby("category")[["market_price", "sale_price"]].agg(np.mean).sort_values(by = "sale_price", ascending = False).plot(kind = "bar")
plt.xlabel("Categories")
plt.xticks(rotation = 45)
plt.ylabel("Average Price")
plt.title("Market vs Sales Price Average")
plt.show()
```

```python
df.pivot_table(values = ["market_price", "sale_price"], index= "category", aggfunc=[np.min, np.max, np.mean])
```

```python
gdp_pivot = df.pivot_table(values = ["market_price", "sale_price"], index = ["category", "sub_category"], aggfunc=[np.min, np.max, np.mean])
gdp_pivot
```

```python
gdp_pivot.loc["Cleaning & Household", "mean"].plot(kind = "bar")
plt.xticks(rotation = 45)
plt.title("Market vs Sales Price (Average) for Sub categories")
plt.xlabel("Sub Categories")
plt.ylabel("Average Price")
plt.show()
```
