## Importing Libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

## Importing Data and EDA 

```python
df = pd.read_csv("D:/Data Camp/Data Scientist with Python/3. Joining Data with Pandas/BigBasketData/BigBasket Products.csv")
df
```

```python
df.drop(["index"], axis = 1, inplace = True, errors= "ignore")
```

```python
unrated_df = df[df["rating"].isna()]
unrated_df
```

```python
plt.figure(figsize=(18, 8))
sns.countplot(data = unrated_df, x = "category")
plt.xlabel("Product Category")
plt.ylabel("Count of Products")
plt.xticks(rotation = 45)
plt.title("Product Categories missing ratings")
```

```python
unrated_df.category.value_counts()
```

```python
print(df["category"].unique())
print(len(df["sub_category"].unique()))
```

```python
df.sort_values("sale_price", ascending= False)
```

```python
plt.figure(figsize = (8, 6))
df.groupby("category")[["market_price", "sale_price"]].agg(np.mean).sort_values(by = "sale_price", ascending = False).plot(kind = "bar")
plt.xlabel("Categories")
plt.ylabel("Average Price")
plt.show()
```

```python
df.pivot_table(values = ["market_price", "sale_price"], index= "category", aggfunc=[np.min, np.max, np.mean])
```

```python
sns.boxplot(data = df, y = "sale_price")
```
